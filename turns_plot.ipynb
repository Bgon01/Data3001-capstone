```python
left_df = pd.read_csv("data/f1sim-ref-left.csv")
right_df = pd.read_csv("data/f1sim-ref-right.csv")
line_df = pd.read_csv("data/f1sim-ref-line.csv")
```

```python
left_df.compare(right_df)
```

```python
fig = plt.figure(figsize = (10, 10))
ax = plt.axes(projection = '3d')
X1 = left_df["WORLDPOSX"]
Y1 = left_df["WORLDPOSY"]
Z1 = left_df["WORLDPOSZ"]
ax.scatter(X1,Y1,Z1, c='green')
X2 = right_df["WORLDPOSX"]
Y2 = right_df["WORLDPOSY"]
Z2 = right_df["WORLDPOSZ"]
ax.scatter(X2,Y2,Z2, c='red')
X = line_df["WORLDPOSX"]
Y = line_df["WORLDPOSY"]
Z = line_df["WORLDPOSZ"]
ax.scatter(X,Y,Z, c='blue')
ax.set_zlim(0, 200)
plt.show
```
