```python
coefficients = np.polyfit(left_df["WORLDPOSX"], left_df["WORLDPOSY"], 3)
```

```python
equation = f"{coefficients[0]}x^3 + {coefficients[1]}x^2 + {coefficients[2]}x + {coefficients[3]}"
```

```python
x_fit = np.linspace(min(left_df["WORLDPOSX"]), max(left_df["WORLDPOSX"]), 100)
y_fit = coefficients[0] * x_fit**3 + coefficients[1] * x_fit**2 + coefficients[2] * x_fit + coefficients[3]
plt.plot(x_fit, y_fit, label=f'Fitted Curve: {equation}', color='red')
```
